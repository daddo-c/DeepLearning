Classificação de sons em diferentes classes pré definidas

```python
#Importação de bibliotecas utilizadas

import tensorflow as tf
import numpy as np
import librosa
from IPython.display import display, Audio
from google.colab import files
import warnings

warnings.filterwarnings("ignore")
```

```python
#Função de classificação do áudio
def classificar_audio(model_path='model.tflite', audio_path='audio.mp3', sr=16000):
    """
    Classificar um arquivo de áudio usando um modelo TFLite do Teachable Machine (som)

    Retorno:
      - predicao: nome da classe prevista
    """
    # Carregar o modelo TFLite
    interpreter = tf.lite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    # Descobrir número de amostras que o modelo espera
    input_shape = input_details[0]['shape']  # ex: [1, 44032]
    n_amostras_esperadas = input_shape[1]

    # Carregar o áudio e ajustar tamanho
    y, _ = librosa.load(audio_path, sr=sr, mono=True)
    if len(y) > n_amostras_esperadas:
        y = y[:n_amostras_esperadas]
    else:
        y = np.pad(y, (0, max(0, n_amostras_esperadas - len(y))), mode='constant')

    # Preparar batch
    input_data = np.expand_dims(y, axis=0).astype(np.float32)

    # Colocar tensor e executar inferência
    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()
    output_data = interpreter.get_tensor(output_details[0]['index'])

    # Mapear classes
    classes = ['Background noise', 'Natureza', 'Cidade', 'Metal']
    predicao = classes[np.argmax(output_data)]

    return predicao
```

Execução do programa:

```python
# Upload do áudio
uploaded = files.upload()
audio_path = list(uploaded.keys())[0]

# Reproduzir áudio
display(Audio(audio_path, autoplay=True))

# Classificar
predicao = classificar_audio('model.tflite', audio_path)
print(f"Classe detectada: {predicao}")


```
